```python
import pyreadstat
import pandas as pd
from pathlib import Path

# 👇 CORRECTO
BASE = Path("../data/raw")

archivos = {
    "enaho_2010":     BASE / "enaho/ENAHO 2010.sav",
    "enaho_2011":     BASE / "enaho/ENAHO 2011.sav",
    "enaho_2012":     BASE / "enaho/ENAHO 2012.sav",
    "enaho_2015":     BASE / "enaho/ENAHO 2015.sav",
    "enaho_2018":     BASE / "enaho/ENAHO 2018.sav",
    "enaho_2020":     BASE / "enaho/ENAHO 2020.sav",
    "enaho_2021":     BASE / "enaho/ENAHO 2021.sav",
    "enaho_2022":     BASE / "enaho/BdBasePublica 2022.sav",
    "enaho_2023":     BASE / "enaho/GPES-ELAB-GEBD-ENAHO-2023_BdBasePublica.sav",
    "enaho_2024":     BASE / "enaho/GPES-ELAB-GEBD-ENAHO-2024_BdPublica.sav",
    "enaho_2025":     BASE / "enaho/GPES-ELAB-GEBD-ENAHO-2025_BdPublica.sav",
    "enigh_gastos":   BASE / "enigh/GPES-ELAB-GEBD-ENIGH-2024A_BdPublicaGastos.sav",
    "enigh_hogares":  BASE / "enigh/GPES-ELAB-GEBD-ENIGH-2024A_BdPublicaHogares.sav",
    "enigh_personas": BASE / "enigh/GPES-ELAB-GEBD-ENIGH-2024A_BdPublicaPersonas.sav",
}

resumen = []

for nombre, ruta in archivos.items():
    try:
        _, meta = pyreadstat.read_sav(str(ruta), row_limit=0)
        primera_col = meta.column_names[0]
        cols_total  = len(meta.column_names)

        df, _ = pyreadstat.read_sav(str(ruta), usecols=[primera_col])
        filas = len(df)

        resumen.append({
            "archivo":       nombre,
            "filas":         filas,
            "columnas":      cols_total,
            "primera_col":   primera_col,
        })
        print(f"✅ {nombre:<20}: {filas:>8,} filas × {cols_total:>4} cols")

    except Exception as e:
        print(f"❌ {nombre}: {e}")

df_resumen = pd.DataFrame(resumen)
total = df_resumen["filas"].sum()

print(f"\n{'─'*55}")
print(f"  TOTAL FILAS brutas: {total:>12,}")
```

```python
# Ver columnas clave de enaho_2025 y enigh_gastos
for nombre, ruta in [
    ("enaho_2025",    BASE / "enaho/GPES-ELAB-GEBD-ENAHO-2025_BdPublica.sav"),
    ("enigh_gastos",  BASE / "enigh/GPES-ELAB-GEBD-ENIGH-2024A_BdPublicaGastos.sav"),
]:
    _, meta = pyreadstat.read_sav(str(ruta), row_limit=0)
    print(f"\n{'='*55}")
    print(f"ARCHIVO: {nombre} — {len(meta.column_names)} columnas")
    print(f"\nNombres de columnas:")
    for i, (col, label) in enumerate(zip(meta.column_names, meta.column_labels)):
        print(f"  {i:>3}. {col:<25} | {label}")
    if i > 50:
        print(f"  ... y {len(meta.column_names)-50} columnas más")
```

```python
# Muestra de datos reales — columnas clave
import pandas as pd
import pyreadstat
from pathlib import Path

BASE = Path("data/raw")

# ── Enaho 2025: solo columnas clave ──────────────────────
cols_enaho = [
    "FACTOR", "ID_HOGAR", "REGION", "ZONA",
    "TamHog",                          # tamaño del hogar
    "A4", "A5",                        # sexo, edad
    "NivInst",                         # nivel instrucción
    "CondAct",                         # condición actividad (ocupado/desocupado)
    "ipcb", "ipcn",                    # ingreso per cápita bruto/neto
    "ithb", "ithn",                    # ingreso total hogar bruto/neto
    "cba",                             # valor canasta básica alimentaria
    "lp",                              # línea de pobreza
    "np",                              # nivel de pobreza
    "Q_IPCN", "D_IPCN",               # quintil y decil
    "IPM_Pobreza",                     # pobreza multidimensional
]

df_enaho, _ = pyreadstat.read_sav(
    str(BASE / "enaho/GPES-ELAB-GEBD-ENAHO-2025_BdPublica.sav"),
    usecols=cols_enaho
)

print(f"Enaho 2025: {len(df_enaho):,} filas × {len(df_enaho.columns)} cols seleccionadas")
print(df_enaho.head(3).to_string())
print(f"\nNulos por columna:")
print(df_enaho.isnull().sum())
print(f"\nEstadísticas de ingreso y canasta:")
print(df_enaho[["ipcb","ithb","cba","lp"]].describe().round(0))
```

```python
# ── ENIGH Gastos: muestra completa (solo 21 cols) ────────
df_gastos, _ = pyreadstat.read_sav(
    str(BASE / "enigh/GPES-ELAB-GEBD-ENIGH-2024A_BdPublicaGastos.sav")
)

print(f"\nENIGH Gastos: {len(df_gastos):,} filas × {len(df_gastos.columns)} cols")
print(df_gastos.head(5).to_string())
print(f"\nDivisiones de gasto únicas (CCIF 2018):")
print(df_gastos[["CCIF18_2DIG","DES_CCIF18_2DIG"]].drop_duplicates().sort_values("CCIF18_2DIG").to_string())
print(f"\nGasto mensualizado — estadísticas:")
print(df_gastos["GASTO_MENS"].describe().round(0))
```

```python
# ════════════════════════════════════════════════════════
# BLOQUE 10 — GRÁFICO 1: ingreso vs costo de la canasta
# Investigación: presión económica por año
# ════════════════════════════════════════════════════════

import pandas as pd
import numpy as np
import plotly.graph_objects as go
from pathlib import Path

DW = Path("data/warehouse")

# Usar el df que ya tienes cargado; si no existe, lo vuelve a cargar
try:
    df
except NameError:
    df = pd.read_parquet(DW / "fact_personas.parquet")

# Validación mínima
columnas_necesarias = ["anio", "ingreso_percapita_bruto", "valor_cba"]
faltantes = [c for c in columnas_necesarias if c not in df.columns]
if faltantes:
    raise ValueError(f"Faltan columnas para el gráfico: {faltantes}")

# Si no existe la razón canasta/ingreso, se calcula
if "ratio_cba_ingreso" not in df.columns:
    df["ratio_cba_ingreso"] = (
        df["valor_cba"] / df["ingreso_percapita_bruto"]
    ).replace([np.inf, -np.inf], np.nan)

# Limpiar valores inválidos
df_g1 = df[["anio", "ingreso_percapita_bruto", "valor_cba", "ratio_cba_ingreso"]].copy()
df_g1 = df_g1.dropna()
df_g1 = df_g1[
    (df_g1["ingreso_percapita_bruto"] > 0) &
    (df_g1["valor_cba"] > 0)
]

# Resumen anual
resumen_anual = (
    df_g1.groupby("anio", as_index=False)
    .agg(
        ingreso_percapita_promedio=("ingreso_percapita_bruto", "mean"),
        cba_promedio=("valor_cba", "mean"),
        porcentaje_ingreso_en_cba=("ratio_cba_ingreso", "mean")
    )
    .sort_values("anio")
)

# Convertir a porcentaje
resumen_anual["porcentaje_ingreso_en_cba"] = resumen_anual["porcentaje_ingreso_en_cba"] * 100

print("Resumen anual del gráfico 1:")
print(resumen_anual.round(2).to_string(index=False))

# Guardar resumen por si luego lo ocupas para dashboard web
resumen_anual.to_csv(DW / "grafico_ingreso_vs_cba.csv", index=False)

# Gráfico interactivo
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=resumen_anual["anio"],
    y=resumen_anual["ingreso_percapita_promedio"],
    mode="lines+markers",
    name="Ingreso per cápita promedio",
    hovertemplate="Año: %{x}<br>Ingreso per cápita: ₡%{y:,.0f}<extra></extra>"
))

fig.add_trace(go.Scatter(
    x=resumen_anual["anio"],
    y=resumen_anual["cba_promedio"],
    mode="lines+markers",
    name="Valor promedio CBA",
    hovertemplate="Año: %{x}<br>Canasta básica: ₡%{y:,.0f}<extra></extra>"
))

fig.add_trace(go.Scatter(
    x=resumen_anual["anio"],
    y=resumen_anual["porcentaje_ingreso_en_cba"],
    mode="lines+markers",
    name="% del ingreso destinado a CBA",
    yaxis="y2",
    hovertemplate="Año: %{x}<br>% ingreso en CBA: %{y:.1f}%<extra></extra>"
))

fig.update_layout(
    title="Ingreso per cápita vs costo de la canasta básica por año",
    xaxis_title="Año",
    yaxis_title="Colones (₡)",
    yaxis2=dict(
        title="% del ingreso per cápita",
        overlaying="y",
        side="right"
    ),
    template="plotly_white",
    hovermode="x unified",
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5
    ),
    width=1100,
    height=600
)

fig.show()

# Guardar versión interactiva para abrir en navegador luego
fig.write_html(DW / "grafico_ingreso_vs_cba.html")

print("\n✅ Gráfico 1 generado")
print("✅ CSV guardado en: data/warehouse/grafico_ingreso_vs_cba.csv")
print("✅ HTML interactivo guardado en: data/warehouse/grafico_ingreso_vs_cba.html")
```
